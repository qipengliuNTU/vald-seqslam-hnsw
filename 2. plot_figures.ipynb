## Outline

1. PR curveï¼š For each dataset, plot four PR curves-- seqslam, seqslam+dbow, seqslam+vlad, seqslam+vlad+hnsw

2. Search time: For each dataset, plot two curves -- seqslam+vlad, seqslam+vlad+hnsw

```python
import pickle
import matplotlib.pyplot as plt
```

### 1. PR curves

#### NewCollege

```python
# set all pickle files for NewCollege dataset
nc_pr_seq_file = "../0.dbow+vlad+seqslam/3_seqslam/pickles_for_plot/newcollege_first_seq.pkl"
nc_pr_seq_dbow_file = "../0.dbow+vlad+seqslam/3_seqslam/pickles_for_plot/newcollege_first_dbow.pkl"
nc_pr_seq_vlad_file = "newcollege/4_PR_curve.pickle"
nc_pr_seq_vlad_hnsw_file = "newcollege/4_PR_curve(hnsw).pickle"
```

```python
# load all pickle files for NewCollege dataset

with open(nc_pr_seq_file,'rb') as f1:  
    nc_pr_seq = pickle.load(f1)

with open(nc_pr_seq_dbow_file,'rb') as f2:  
    nc_pr_seq_dbow = pickle.load(f2)
    
with open(nc_pr_seq_vlad_file,'rb') as f3:  
    nc_pr_seq_vlad = pickle.load(f3)

with open(nc_pr_seq_vlad_hnsw_file,'rb') as f4:  
    nc_pr_seq_vlad_hnsw = pickle.load(f4)
```

```python
plt.plot(nc_pr_seq[:, 1], nc_pr_seq[:, 0], '-')
plt.plot(nc_pr_seq_dbow[:, 1], nc_pr_seq_dbow[:, 0], '--')
plt.plot(nc_pr_seq_vlad[:, 1], nc_pr_seq_vlad[:, 0], '-.')
plt.plot(nc_pr_seq_vlad_hnsw[:, 1], nc_pr_seq_vlad_hnsw[:, 0], ':')

plt.title('PR Curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.axis([0, 1.05, 0, 1.05])
plt.grid()
plt.legend(['seq', 'seq+bow', 'seq+vlad', 'seq+vlad+hnsw'], loc="lower right")
plt.show()
```

### 2. Search time

#### NewCollege

```python
# set file path
nc_time_seq_vlad_file = "newcollege/5_timing.pickle"
nc_time_seq_vlad_hnsw_file = "newcollege/5_timing(hnsw).pickle"
```

```python
# load pickle file
with open(nc_time_seq_vlad_file,'rb') as f1:  
    nc_time_seq_vlad = pickle.load(f1)

with open(nc_time_seq_vlad_hnsw_file,'rb') as f2:  
    nc_time_seq_vlad_hnsw = pickle.load(f2)
```

```python
plt.plot(nc_time_seq_vlad[::150, 0], nc_time_seq_vlad[::150, 1], '-', linewidth=2.0)
plt.plot(nc_time_seq_vlad_hnsw[::150, 0], nc_time_seq_vlad_hnsw[::150, 1], '--', linewidth=2.0)

plt.title('Searching time')
plt.xlabel('Number of images')
plt.ylabel('Time')
plt.grid()
plt.legend(['seq+vlad', 'seq+vlad+hnsw'], loc="right")
plt.show()
```
